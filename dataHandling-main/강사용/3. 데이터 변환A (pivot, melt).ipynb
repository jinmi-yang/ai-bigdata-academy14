```python
import pandas as pd

df = pd.DataFrame({
    "index": [1,1,1,2,2,2],
    "column": ["col_1", "col_2", "col_3", "col_1", "col_2", "col_3"],
    "value_1": [1, 2, 3, 4, 5, 6],
    "value_2": ["x", "y", "z", "q", "w", "t"]})
df

# 이해를 돕기 위한 대체 df
# df = pd.DataFrame({
#     "index": ["남", "남", "남", "여", "여", "여"],
#     "column": ["10대", "20대", "30대", "10대", "20대", "30대"],
#     "value_1": ["170Cm", "173Cm", "172Cm", "160Cm", "163Cm", "162Cm"],
#     "value_2": ["68Kg", "70Kg", "80Kg", "55Kg", "53Kg", "55Kg"]})
# df
```

# 2. pivot을 이용한 변환 

### pivot: index, column, values를 지정하여 pivot

```python
df.pivot(index="index", columns="column", values="value_1")
```

### 동일한 기능 다른 표현

```python
df.pivot(index="index", columns="column")["value_1"]
```

### 복수의 pivot column 지정

```python
df.pivot(index="index", columns="column", values=["value_1", "value_2"])
```

### pivot column 미지정

```python
df.pivot(index="index", columns="column")
```

### index 미지정 시 변환되는 데이터 구조 주의 

```python
df.pivot(columns="column", values="value_1") #의도치 않은 결과가 나오므로 주의할 것
```

# 3. melt를 이용한 변환

### melt: id_vars를 제외한 나머지 컬럼을 풀어서 쌓음

```python
df.melt(id_vars=["index", "column"])
```

### id 및 value 컬럼 지정

```python
df.melt(id_vars=["index", "column"], value_vars=["value_1"])
```
