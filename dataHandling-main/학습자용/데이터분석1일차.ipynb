## 1. 연습 데이터 생성 및 concat

### df1, df2, df3 데이터 생성

```python
df1 = pd.DataFrame({
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"]},
        index=[0, 1, 2, 3])

df2 = pd.DataFrame({
        "E": ["E4", "E5", "E6", "E7"],
        "F": ["F4", "F5", "F6", "F7"],
        "G": ["G4", "G5", "G6", "G7"],
        "H": ["H4", "H5", "H6", "H7"]},
        index=[0, 1, 2, 3])

df3 = pd.DataFrame({
        "I": ["I8", "I9", "I10", "I11"],
        "J": ["J8", "J9", "J10", "J11"],
        "K": ["K8", "K9", "K10", "K11"],
        "L": ["L8", "L9", "L10", "L11"]},
        index=[0, 1, 2, 3])

result = pd.concat([df1, df2, df3], axis=1) #concat(axis=1)해야 옆(열)으로 붙는다. default값은 axis=0
result # concat 결과 데이터 result 확인
```

## 2. concat의 parameter 변경

### axis = 0: 가로 결합이 아닌 세로 결합

```python
result = pd.concat([df1, df2, df3]) #axis=0이 default값
result 
```

# 종합실습

```python
pop_seoul = pd.read_csv('서울인구통계.txt', sep='\t', engine='python', thousands=',', encoding='utf-8')
pop_seoul.head()
```

```python
pop1 =pop_seoul[['자치구', '인구', '인구.3', '인구.6', '65세이상고령자']] #필요한 자료만 pop
pop_seoul=pop1.drop([0,1]) #필요한 자료에서 자료가 아닌 head값 제외 drop (drop은 지정한 column명 모두 삭제, 원본 변경)

#columns명 새로 다시 정의, inplace=True로 원본 변경
pop_seoul.rename(columns={pop_seoul.columns[0]:'구별', pop_seoul.columns[1]:'인구수', pop_seoul.columns[2]:'한국인'
                          , pop_seoul.columns[3]:'외국인', pop_seoul.columns[4]:'고령자'},inplace = True)
pop_seoul.head()
```

```python
cctv_seoul = pd.read_excel('서울시CCTV.xlsx') #데이터 불러오기
cctv_seoul.head() #데이터 개요 파악
```

```python
#columns명 새로 다시 정의, inplace=True로 원본 변경

cctv_seoul.rename(columns={cctv_seoul.columns[0]: '구별', cctv_seoul.columns[1]: 'cctv수'}, inplace=True )
cctv_seoul.head() #데이터 재확인

#cctv수 column의 값을 기준으로 오름차순 정리 (ascending=True) 후 데이터 확인
cctv_seoul.sort_values(by='cctv수', ascending=True).head(10) 
```

```python
#cctv수 column의 값을 기준으로 내림차순 정리 (ascending=False) 후 데이터 확인
cctv_seoul.sort_values(by='cctv수', ascending=False).head(10)
```

```python
cctv_seoul.isnull().sum() #cctv_seoul의 데이터에 null값을 true(1), null이 아니면 false(0)으로 sum()하면 true(null)의 총 개수 도출

#cctv_seoul의 2011년 이전 column에 null값을 true로 바꾸고, true값을 인덱스에 넣음으로써, cctv_seoul자료에서 null값이 있는 행만 가져옴
cctv_seoul.loc[cctv_seoul.isnull()['2011년 이전']] 
#null값을 대체, 유지, 삭제 중에 결정해야 함.
```

```python
#18,17,16년을 제외한 해에 대비 18,17,16년의 cctv수를 가지고 최근증가율을 정의하여, cctv_seoul에 새 column으로 삽입
cctv_seoul['최근증가율'] = ((cctv_seoul['2018년']+cctv_seoul['2017년']+cctv_seoul['2016년'])/(cctv_seoul['cctv수']-cctv_seoul['2018년']+cctv_seoul['2017년']+cctv_seoul['2016년']))*100

#최근증가율 column의 value값으로 내림차순 정리
cctv_seoul.sort_values(by='최근증가율', ascending=False).head() 
```

```python
#다시 서울시 인구수 데이터 파악
pop_seoul.head() #1행인 합계는 불필요
pop_seoul.drop([2], inplace=True) #index명이 2인 행 제외, 원본 변경
pop_seoul.head() #데이터 재확인
```

```python
pop_seoul.dtypes #데이터의 타입 확인
pop_seoul = pop_seoul.astype({'인구수':'int', '한국인':'int', '외국인':'int', '고령자':'int'}) #데이터 타입 object에서 int형으로 변경
pop_seoul['외국인 비율'] = pop_seoul['외국인']/pop_seoul['인구수']*100 #데이터 수학적 계산 가능 (int형)
pop_seoul['고령자 비율'] = pop_seoul['고령자']/pop_seoul['인구수']*100 #비율 확인 
pop_seoul.head()
```

```python
#외국인 비율을 높은 구 기준으로 내림차순 확인

pop_seoul.sort_values(by='외국인 비율', ascending=False).head() 
```

```python
#고령자 비율을 높은 구 기준으로 내림차순 확인

pop_seoul.sort_values(by='고령자 비율', ascending=False).head()
```
